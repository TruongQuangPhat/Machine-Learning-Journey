## What are Recommender Systems
- Recommender Systems are tools that suggest items to users based on their behaviour, preferences or past interactions. They help users find relevant products, movies, songs or content without manually searching for them. They are widely used in platforms like Youtube, Amazon and Netflix.
    - Suggest items automatically
    - Use user behaviour and patterns.
    - Improve content discovery
    - Common in e-commerce and entertainment apps

## Approaches in Recommender Systems

For the systems to understand and provide recommendations, there are several different approaches, evolving from basic algorithm to complex neural network architectures. Below are the most common categories of models:

**1. Content-Based Filtering**

This model focuses on analyzing the features of the item itself and user's historical profile.
- **How it works:** It extracts item information (e.g., genre, director, descriptive keywords) into vectors. When a user positively interacts with an item, the system finds other items with similar feature vectors to recommend.
- **Pros:** Handles the "new item" problem well, as it only needs metadata to analyze. It operates independently without needing to collect data from other users.
- **Cons:** Highly prone to the "filter bubble" effect. The system tends to suggest content that is too safe and familiar, lacking the ability to broaden the user's interests.

**2. Neighborhood-based Collaborative Filtering (NBCF)**

In contrast to Content-Based Filtering, this method completely ignores the item's internal content. Instead, it forcuses on exploiting the wisdom of the crowd through user interaction behaviour (sush as clicks, ratings, and purchases).
- **User-based CF:** Finds users with similar interaction histories. If User A and User B share similar tastes, the system will recommend an item that A just liked to B.
- **Item-based CF:** Provides recommendations based on items that are frequently consumed together by many different users.
- **Pros:** Brings serendipity, helping users discover new and interesting items they might not have actively thought of.
- **Cons:** Faces a significant hurdle with the "cold-start" problem. The system struggles completely when dealing with a brand-new user or a product that no one has interacted with yet.

**3. Matrix Factorization Collaborative Filtering**

This is a core techniques used to solve the problem of massive but highly sparse User-Item interaction matrices found in the real world.
- **How it works:** It uses linear algebra to decompose the original, large matrix into two smaller, lower-dimensional matrices representing the User and the Item. This process helps discover "latent features" that humans cannot directly define or observe.
- **Highlights:** Offers fast computation and the ability to accurately predict the empty cells (missing ratings) in the matrix. This is the foundational algorithm that drove the success of the famous Netflix Prize competition.

**4. Deep Learning Models**

When the data scale reaches millions of users and billions of interactions, Deep Learning models are applied to learn highly complex, non-linear relationships.
- **Neural Collaborative Filtering (NCF):** Replaces basic matrix multiplication (dot product) with neural network layers (like Multi-Layer Perceptrons), allowing the model to be much more flexible in learning representations.
- **Two-Tower Model:**  An architecture consisting of two independent neural network branches used to embed User and Item information into the same vector space. This is the industry standard for the "retrieval" stage at a massive scale.
- **Sequential Models:** Treats the user's interaction history as a sequence of events over time. It applies architectures like RNNs or Transformers to predict the user's next action based on their current context.

## Recommender Systems Models

### Content-Based Filtering
**1. Model Construction**

A Content-Based Filtering system is built on two main pillars: the Item Profile and the User Profile.

- **Step 1:** Feature Extraction to create the Item Profile.

  Each item must be transformed into a mathematical vector $x^{(i)}$. In practice, items do not come with pre-calculated scores like a romance level of 0.9. These features must be constructed from raw data such as metadata, text, or images. Particularly for systems recommending articles or documents, applying text mining techniques is essential. Common methods include using **TF-IDF** or **CountVectorizer** to turn product descriptions into vocabulary vectors, utilizing **Word Embeddings** like **Word2Vec** or **BERT** to capture deeper semantic meaning, and applying one-hot encoding for categorical features like genre, director, or producer.

- **Step 2:** Building the User Profile.

  Each user is represented by a parameter vector, such as $w^{(j)}$, and a bias $b^{(j)}$. This vector has the same dimensionality as the item profile. It represents the weight or level of importance that user $j$ places on each specific feature of the items.

- **Step 3:** Prediction.

  The predicted score $\hat{y}^{(i,j)}$ that user $j$ would give to item $i$ is calculated as the dot product of the user's preference vector and the item's feature vector:$$\hat{y}^{(i,j)} = w^{(j)} \cdot x^{(i)} + b^{(j)}$$

**2. Loss Function**

The choice of loss function depends on the type of user feedback data available.

- **Case 1:** Explicit Feedback (e.g., 1-5 star ratings)

  This is treated as a regression problem. A suitable loss function is Mean Squared Error (MSE) combined with L2 Regularization to prevent overfitting and keep the weights $w$ from becoming too large. The loss function for user $j$ is:
  $$
  J(w^{(j)}, b^{(j)}) = \frac{1}{2} \sum_{i:r(i, j)=1} \left(w^{(j)} \cdot x^{(i)} + b^{(j)} - y^{(i, j)} \right)^2 + \frac{\lambda}{2} \sum_{k=1}^{n}\left(w_{k}^{(j)} \right)^2
  $$

- **Case 2:** Implicit Feedback (e.g., click/no click, buy/no buy)
  
  In this scenario, the task becomes a binary classification problem. The predicted score is passed through a Sigmoid function $\sigma(z) = \frac{1}{1+e^{-z}}$, and the loss function used is Binary Cross-Entropy (Log Loss):
  $$
  J(w^{(j)}, b^{(j)}) = - \sum_{i:r(i, j)=1} \left[y^{(i, j)}\log(\hat{y}^{(i, j)}) + (1 - \hat{y}^{(i, j)})\log(1 - \hat{y}^{(i, j)}) \right] + \frac{1}{2}\sum_{k=1}^{n}\left(w_{k}^{(j)} \right)^2
  $$

**3. Optimization**

  To find the weights $w^{(j)}$ and bias $b^{(j)}$ that minimize the loss function, Gradient Descent is commonly used. The algorithm calculates the partial derivative of the loss function with respect to each parameter and updates them iteratively. The weight update rule for the regression problem is:
  $$
  w_{k}^{(j)} := w_{k}^{(j)} - \alpha \left(\sum_{i:r(i, j)=1} \left(\hat{y}^{(i, j)} - y^{(i, j)}\right) x_{k}^{(i)} + \lambda w_{k}^{(j)}\right)
  $$
  Here, $\alpha$ represents the learning rate. In production environments, engineers often rely on more advanced optimizers like **Adam** or **RMSprop** for deep learning models, or direct solvers like L-BFGS for faster convergence, rather than implementing basic Gradient Descent from scratch.

**4. Evaluation:** Pros and Cons

Understanding the strengths and limitations of this model helps determine the appropriate use cases.

- Pros:
  - No item cold-start problem: When a new item is added to the system and its features are extracted, it can be recommended immediately to relevant users without needing prior interactions.
  - User independence: The algorithm learns solely from a specific user's history and does not require data from other users on the platform.
  - High interpretability: It is straightforward to explain recommendations, such as informing a user that an article was suggested because they frequently read about certain topics.
- Cons:
  - Heavy reliance on feature engineering: The model's accuracy depends entirely on the quality of feature extraction. If only basic metadata is available, the recommendations will not be highly relevant.
  - Filter bubble effect: The system tends to suggest items that are very similar to what the user has already consumed. It lacks the ability to introduce entirely new genres or categories, limiting serendipity.
  - User cold-start problem: If a new user registers without any interaction history, the model cannot build a profile and is unable to generate recommendations.

#### Code Example: Content-Based Filtering – MovieLens 100K

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import Ridge
from math import sqrt
```

```python
# Load data
def load_data(path="data/ml-100k"):

    # Load user data
    u_cols = ["user_id", "age", "sex", "occupation", "zip_code"]
    users = pd.read_csv(f"{path}/u.user", sep="|", names=u_cols, encoding="latin-1")
    n_users = users.shape[0]

    # Load ratings data
    r_cols = ["user_id", "movie_id", "rating", "unix_timestamp"]
    ratings_base = pd.read_csv(f"{path}/ua.base", sep="\t", names=r_cols, encoding="latin-1")
    ratings_test = pd.read_csv(f"{path}/ua.test", sep="\t", names=r_cols, encoding="latin-1")

    rate_train = ratings_base.values
    rate_test = ratings_test.values

    # Load item data
    i_cols = ['movie_id', 'movie_title', 'release_date', 'video_release_date',
              'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation',
              "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama',
              'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
              'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
    items = pd.read_csv(f"{path}/u.item", sep="|", names=i_cols, encoding="latin-1")
    n_items = items.shape[0]

    return users, items, rate_train, rate_test, n_users, n_items
```

```python
# Build item profiles (TF-IDF)
def build_item_profiles(items):
    X0 = items.values
    gerne_matrix = X0[:, -19:]

    transformer = TfidfTransformer(smooth_idf=True, norm="l2")
    item_profiles = transformer.fit_transform(gerne_matrix.tolist()).toarray()
    return item_profiles
```

```python
# Get items rated by user
def get_items_rated_by_user(rate_matrix, user_id):
    y = rate_matrix[:, 0]
    ids = np.where(y == user_id + 1)[0]
    item_ids = rate_matrix[ids, 1] - 1
    scores = rate_matrix[ids, 2]
    return item_ids.astype(int), scores
    
```

```python
# Train content-based model
def train_content_based_model(item_profiles, rate_train, n_users):
    d = item_profiles.shape[1]
    W = np.zeros((d, n_users))
    b = np.zeros((1, n_users))

    for n in range(n_users):
        ids, scores = get_items_rated_by_user(rate_train, n)
        clf = Ridge(alpha=0.1, fit_intercept=True)

        Xhat = item_profiles[ids, :]
        clf.fit(Xhat, scores)

        W[:, n] = clf.coef_
        b[0, n] = clf.intercept_

        if n % 100 == 0:
            print(f"Trained model for user {n}/{n_users}")
    
    return W, b
```

```python
# Predict ratings for user
def predict(item_profiles, W, b):
    return item_profiles @ W + b
```

```python
# Evaluate model
def evaluate(Yhat, rates, n_users):
    se = 0
    cnt = 0

    for n in range(n_users):
        ids, scores_truth = get_items_rated_by_user(rates, n)
        scores_pred = Yhat[ids, n]

        e = scores_truth  - scores_pred
        se += np.sum(e ** 2)
        cnt += len(ids)
    return np.sqrt(se / cnt)
```

```python
# Recommend function
def recommend_top_k(user_id, Yhat, rate_train, items, k=5   ):
    rated_items, _ = get_items_rated_by_user(rate_train, user_id)

    scores = Yhat[:, user_id]
    scores[rated_items] = -np.inf

    top_ids = np.argsort(scores)[-k:][::-1]

    print(f"\nTop {k} recommendations for User {user_id + 1}:")

    for idx in top_ids:
        print(f"- {items.iloc[idx]['movie_title']} (Predicted: {scores[idx]:.2f})")
```

```python
users, items, rate_train, rate_test, n_users, n_items = load_data()

print("Number of users:", n_users)
print("Number of items:", n_items)

tfidf = build_item_profiles(items)

W, b = train_content_based_model(tfidf, rate_train, n_users)

Yhat = predict(tfidf, W, b)

print("\nRMSE Train:", evaluate(Yhat, rate_train, n_users))
print("RMSE Test :", evaluate(Yhat, rate_test, n_users))

recommend_top_k(user_id=10, Yhat=Yhat, rate_train=rate_train, items=items, k=5)
```

### Neighborhood-based Collaborative Filtering (NBCF)

Unlike Content-Based Filtering recommendation systems that require extracting item features, or Matrix Factorization that requires the machine to self-learn latent factors, NBCF is a memory-based method. This algorithm memorizes the entire interaction matrix containing the user's rating history and directly searches for similarities using the K-Nearest-Neighbors (KNN) algorithm.

**1. Two Model Construction Strategies**

The system exploits the interaction matrix through two core approaches:
- **User-User CF (Finding similar users):** The basic idea is that people with similar tastes to yours will also like similar products. To predict the score you would give a movie, the system finds a group of users with a rating history extremely similar to yours. If that group gives a high score to the movie in question, the system infers that you will like it too. The focus of this method is building a similarity matrix among users.
- **Item-Item CF (Finding similar items):** The idea here is that because you liked a movie, you might like other movies with characteristics that the community rates similarly. The system ignores comparing you to other people. It simply scans the movies you have watched, and if you previously gave a high score to movie A, while the data shows movie A and movie B are frequently rated identically by the community, it infers you will like movie B. The focus of this method is building a similarity matrix among items.

**2. Core Mathematics**

This model has no weight training process using Gradient Descent or a loss function. Its core lies in vector space mathematics to measure distance.

- **Step 1: Data Normalization**

  To eliminate the bias of users who are too strict or too easygoing, the rating score $y_{i,u}$ of user $u$ for item $i$ is normalized by subtracting the average score $\mu_u$ of that specific user:
  $$
  \bar{y}_{i,u} = y_{i,u} - \mu_u
  $$
  Unrated cells are filled with a value of 0. The matrix now contains positive numbers representing preference, negative numbers representing dislike, and 0 for unknown. This allows for storage as a sparse matrix to optimize memory.
  
- **Step 2: Measuring Similarity**
  
  The distance between two normalized vectors is typically calculated using Cosine Similarity:
  $$
  sim(u, v) = \frac{\vec{u} \cdot \vec{v}}{\|\vec{u}\|_2 \|\vec{v}\|_2}
  $$
  The Cosine calculation on mean-centered data is actually the Pearson correlation coefficient. The returned value ranges from -1 to 1, where 1 means completely identical and -1 means completely opposite.

- **Step 3: Prediction Formula**

  The normalized predicted score $\hat{y}_{i,u}$ is the weighted average of the scores from $k$ neighboring entities $N$ that have rated item $i$:
  $$
  \hat{y}_{i,u} = \frac{\sum_{u_j \in N(u,i)} \bar{y}_{i,u_j} sim(u, u_j)}{\sum_{u_j \in N(u,i)} |sim(u, u_j)|}
  $$
  The denominator uses the absolute value so the ratio is not distorted when there are negative similarity values. The final predicted score on the original scale is found by adding back the average score:
  $$
  \hat{r}_{i,u} = \hat{y}_{i,u} + \mu_u
  $$

**3. System Execution: Why Does Practice Favor Item-Item?**

Theoretically, comparing people to people is more natural to conventional thinking. However, when deployed on massive systems, this strategy easily overloads servers. The Item-Item algorithm is typically chosen to solve three core technical problems:
- **Computational complexity:** The number of users is usually thousands of times larger than the number of items. Calculating the matrix for users takes $O(M^2)$ complexity, which is unfeasible. The item-to-item matrix with $O(N^2)$ complexity will be much smaller and lighter.
- **Stability and caching:** User preferences change continuously after every interaction, requiring the system to recalculate in real-time. Meanwhile, the nature of an item is very static because it is shaped by millions of people. Engineers can run the algorithm for items at night and save the results in the cache to serve high-speed retrieval during the day.
- **Data sparsity:** A user's vector is very sparse because one person rarely interacts with many items. Conversely, a popular item contains a lot of ratings. Calculating the similarity between two dense item vectors will yield higher accuracy and better reliability.

**4. Evaluation of Pros and Cons**

**Pros:**
- High interpretability, making it very easy to explain to users why they receive specific recommendations.
- Does not rely on barriers of natural language processing or extracting item features.
- Instant reaction when there is a new rating without needing to rerun the entire process from the beginning.

**Cons:**
- Faces the cold-start problem, becoming completely paralyzed by a brand new user or a newly launched item.
- Limited connectivity when data is too sparse; if two users do not coincidentally watch any shared items, the similarity instantly becomes 0.
- Scanning and calculating on the matrix in real-time consumes more hardware resources compared to deep learning models.

**Appendix: Detailed Calculation Illustration**

To understand how the matrix operates, below is how to predict the score user 1 gives to item 1 with the neighborhood size $k = 2$.

**Example 1: Calculation based on User-User CF** 
1. Find neighboring points by examining those who have rated item 1, including users 0, 3, and 5. User 1's similarity to them is 0.83, -0.40, and -0.23 respectively. We select the highest 2, which are users 0 and 5.
2. Take the normalized scores that users 0 and 5 gave to item 1, which are 0.75 and 0.5 respectively.
3. Calculate the weighted average:$$\hat{y}_{1, 1} = \frac{0.83 \times 0.75 + (-0.23) \times 0.5}{0.83 + |-0.23|} \approx 0.48$$
4. Return the actual score by taking the average score of user 1, which is 2.75, and adding 0.48, resulting in 3.23.
 
**Example 2: Calculation based on Item-Item CF** 
1. Find neighboring points by examining the items that user 1 has rated, including items 0, 2, 3, and 4. Item 1's similarity to these items is 0.77, 0, -0.64, and -0.14 respectively. We select the highest 2 items, which are items 0 and 2.
2. Take the normalized scores that user 1 gave to these 2 neighboring items, with item 0 being 2.4 and item 2 being 2.25.
3. Calculate the weighted average:$$\hat{y}_{1, 1} = \frac{0.77 \times 2.4 + 0 \times 2.25}{|0.77| + |0|} = \frac{1.848}{0.77} = 2.4$$

#### Code Example: Neighborhood-based Collaborative Filtering – MovieLens 100K

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
```

```python
class CF(object):
    def __init__(self, Y_data, k, dist_func=cosine_similarity, uuCF=1):
        self.uuCF = uuCF
        self.Y_data = Y_data if uuCF else Y_data[:, [1, 0, 2]]
        self.k = k
        self.dist_func = dist_func
        self.Ybar_data = None
        # number of users and items. Remember to add 1 since id starts from 0
        self.n_users = int(np.max(self.Y_data[:, 0])) + 1
        self.n_items = int(np.max(self.Y_data[:, 1])) + 1

    def add(self, new_data):
        """
        Update Y_data matrix when new ratings come.
        For simplicity, suppose that there is no new user or item.
        """
        self.Y_data = np.concatenate((self.Y_data, new_data), axis=0)

    def normalize_Y(self):
        users = self.Y_data[:, 0]
        self.Ybar_data = self.Y_data.copy()
        self.mu = np.zeros((self.n_users,))
        for n in range(self.n_users):
            # row indices of rating done by user n
            # since indices need to be integers, we need to convert
            ids = np.where(users == n)[0].astype(np.int32)
            # indices of all ratings associated with user n
            item_ids = self.Y_data[ids, 1]

            # and the corresponding ratings 
            ratings = self.Y_data[ids, 2]

            if len(ratings) == 0:
                self.mu[n] = 0
                continue

            m = np.mean(ratings) 
            # normalize
            self.mu[n] = m
            self.Ybar_data[ids, 2] = ratings - m

        self.Ybar = sparse.coo_matrix((self.Ybar_data[:, 2], (self.Ybar_data[:, 1], self.Ybar_data[:, 0])), shape=(self.n_items, self.n_users))
        self.Ybar = self.Ybar.tocsr()
    
    def similarity(self):
        self.S = self.dist_func(self.Ybar.T, self.Ybar.T)

    def refresh(self):
        self.normalize_Y()
        self.similarity()

    def fit(self):
        self.refresh()

    def __pred(self, u, i, normalized=1):
        """ 
        predict the rating of user u for item i (normalized)
        if you need the un
        """
        # Step 1: find all users who rated i
        ids = np.where(self.Y_data[:, 1] == i)[0].astype(np.int32)

        # Step 2:
        users_rated_i = self.Y_data[ids, 0].astype(np.int32)

        # Step 3: find similarity btw the current user and others who rated i
        sim = self.S[u, users_rated_i]

        # Step 4: find the k most similarity users
        a = np.argsort(sim)[-self.k:]
        # and the corresponding similarity levels
        nearest_s = sim[a]
        r = self.Ybar[i, users_rated_i[a]]
        if normalized:
            # add a small number, for instance, 1e-8, to avoid dividing by 0
            return (r * nearest_s) / (np.abs(nearest_s).sum() + 1e-8)

        return (r * nearest_s) / (np.abs(nearest_s).sum() + 1e-8) + self.mu[u]
    
    def pred(self, u, i, normalized=1):
        """ 
        predict the rating of user u for item i (normalized)
        if you need the un
        """
        if self.uuCF:
            return self.__pred(u, i, normalized)
        else:
            return self.__pred(i, u, normalized)
        
    def recommend(self, u, normalized = 1):
        """
        Determine all items should be recommended for user u. (uuCF =1)
        or all users who might have interest on item u (uuCF = 0)
        The decision is made based on all i such that:
        self.pred(u, i) > 0. Suppose we are considering items which 
        have not been rated by u yet. 
        """
        ids = np.where(self.Y_data[:, 0] == u)[0]
        items_rated_by_u = self.Y_data[ids, 1].tolist()              
        recommended_items = []
        for i in range(self.n_items):
            if i not in items_rated_by_u:
                rating = self.__pred(u, i)
                if rating > 0: 
                    recommended_items.append(i)
        
        return recommended_items 

    def print_recommendation(self):
        """
        print all items which should be recommended for each user 
        """
        print("Recommendation: ")
        for u in range(self.n_users):
            recommended_items = self.recommend(u)
            if self.uuCF:
                print("    Recommend item(s):", recommended_items, "to user", u)
            else: 
                print("    Recommend item", u, "to user(s) : ", recommended_items)
```

```python
r_cols = ["user_id", "movie_id", "rating", "unix_timestamp"]

ratings_base = pd.read_csv('data/ml-100k/ub.base', sep='\t', names=r_cols, encoding='latin-1')
ratings_test = pd.read_csv('data/ml-100k/ub.test', sep='\t', names=r_cols, encoding='latin-1')

rate_train = ratings_base.values
rate_test = ratings_test.values

# indices start from 0
rate_train[:, :2] -= 1
rate_test[:, :2] -= 1
```

```python
rs = CF(rate_train, k = 30, uuCF = 0)
rs.fit()

n_tests = rate_test.shape[0]
SE = 0 # squared error
for n in range(n_tests):
    pred = rs.pred(rate_test[n, 0], rate_test[n, 1], normalized = 0)
    SE += (pred - rate_test[n, 2]) ** 2 

RMSE = np.sqrt(SE / n_tests)
print("Item-Item CF, RMSE =", RMSE)
```

### Matrix Factorization Collaborative Filtering
**1. Mathematical Foundations and Core Concepts**

Before diving into the construction steps, we need to understand the mathematical mechanism that allows this algorithm to create magic from an extremely sparse data matrix (up to 99% empty). This mechanism lies in two concepts:
- **Latent Features ($K$):** Unlike Content-Based filtering where humans manually define explicit labels (e.g., action movies, comedies), Matrix Factorization automatically creates an invisible $K$-dimensional space. The computer doesn't know what to name the first dimension; it only knows that through the optimization process, a group of movies (like superhero movies) consistently scores very high mathematically in this dimension. This process is like listening to a completely new instrumental track: you don't know its official genre, but your brain automatically analyzes and groups it based on rhythm, tone, or instruments. Those invisible analytical dimensions are the Latent Features. The number $K$ (number of latent features) is a hyperparameter decided by the engineer through Cross-Validation.
- **Low-Rank Approximation:** Suppose the original matrix $Y$ has 1 million users and 10,000 movies. Instead of learning all 10 billion data cells, the algorithm actively creates a "bottleneck" by choosing a very small $K$ (e.g., $K=50$). It decomposes the massive matrix into two smaller ones: an item matrix of $M \times K$ and a user matrix of $K \times N$. When multiplying these two matrices to predict, the resulting matrix $\hat{Y}$ has limited degrees of freedom, meaning it has a low rank.
  - **Forced Generalization:** The system cannot memorize discrete scores. It is forced to find the most general underlying patterns governing the entire dataset to fill the empty cells with reasonable numbers.
  - **Denoising:** It discards random, biased ratings and only retains the strongest core preference clusters.
  - **Resource Optimization:** It reduces the problem from processing 10 billion elements to just a few tens of millions of parameters, enabling the system to run smoothly on massive datasets.

**2. Model Construction**

With the $K$-dimensional latent feature space established, the model is built through the following steps:
- **Step 1: Initialize the Item Profile.**
  
  Each item $i$ is represented by a latent feature vector $x^{(i)}$ with $K$ dimensions. Initially, the values in this vector are just random numbers. Through training, the model will automatically adjust these numbers so that they reflect the true nature of the item based on community interaction behavior.

- **Step 2: Initialize the User Profile.**

  Each user $j$ is represented by a latent feature vector $w^{(j)}$ and a bias $b^{(j)}$. The vector $w^{(j)}$ has the same $K$ dimensions as the item vector, representing the degree to which user $j$ cares about each corresponding latent feature. If the user vector and the item vector have large values in the same $K$ dimension, the predicted score will be very high.

- **Step 3: Prediction.**

  - The predicted score $\hat{y}^{(i,j)}$ that user $j$ gives to item $i$ is calculated by the dot product between the user vector and the item vector, plus the user's bias:
  $$
  \hat{y}^{(i,j)} = w^{(j)} \cdot x^{(i)} + b^{(j)}
  $$
  (Note: To thoroughly address natural bias, the actual formula usually adds the item's bias $b^{(i)}$ and the global average score $\mu$).
  
**3. Loss Function**

The choice of loss function depends on the type of user feedback data available.
- **Case 1: Explicit Feedback (e.g., 1-5 star ratings)**
  
  This is a regression problem. The loss function used is Mean Squared Error (MSE) combined with L2 Regularization applied to both user and item parameters to prevent overfitting. The generalized loss function is defined as:
  $$
  J(w, b, x) = \frac{1}{2} \sum_{(i,j): r(i,j)=1} \left(w^{(j)} \cdot x^{(i)} + b^{(j)} - y^{(i,j)} \right)^2 + \frac{\lambda}{2} \sum_{j=1}^{n_u} \sum_{k=1}^K \left(w_k^{(j)} \right)^2 + \frac{\lambda}{2} \sum_{i=1}^{n_m} \sum_{k=1}^K \left(x_k^{(i)} \right)^2
  $$

- **Case 2: Implicit Feedback (e.g., click/no click, buy/no buy)**

  The problem now becomes binary classification. The predicted score is passed through a Sigmoid function $\sigma(z) = \frac{1}{1+e^{-z}}$ to predict the interaction probability, and the loss function shifts to Binary Cross-Entropy (Log Loss) combined with Regularization:
  $$
  J(w, b, x) = - \sum_{(i,j): r(i,j)=1} \left[y^{(i,j)}\log(\sigma(\hat{y}^{(i,j)})) + (1 - y^{(i,j)})\log(1 - \sigma(\hat{y}^{(i,j)})) \right] + \frac{\lambda}{2} \sum_{j=1}^{n_u} \sum_{k=1}^K \left(w_k^{(j)} \right)^2 + \frac{\lambda}{2} \sum_{i=1}^{n_m} \sum_{k=1}^K \left(x_k^{(i)} \right)^2
  $$

**4. Optimization**

To find the weights $w^{(j)}$, biases $b^{(j)}$, and item features $x^{(i)}$ that minimize the loss function, the Gradient Descent algorithm is commonly used. The biggest difference compared to previous models is that $x$ is also a parameter that needs to be optimized simultaneously. The weight update rule (for the Explicit Feedback problem) is:
$$
w_k^{(j)} := w_k^{(j)} - \alpha \left( \sum_{i:r(i,j)=1} \left(\hat{y}^{(i,j)} - y^{(i,j)}\right) x_k^{(i)} + \lambda w_k^{(j)} \right)
$$
$$
b^{(j)} := b^{(j)} - \alpha \sum_{i:r(i,j)=1} \left(\hat{y}^{(i,j)} - y^{(i,j)}\right)
$$
$$
x_k^{(i)} := x_k^{(i)} - \alpha \left( \sum_{j:r(i,j)=1} \left(\hat{y}^{(i,j)} - y^{(i,j)}\right) w_k^{(j)} + \lambda x_k^{(i)} \right)
$$

Here, $\alpha$ represents the learning rate. However, because the loss function of this system is non-convex, using vanilla Gradient Descent can lead to slow convergence. In a production environment, system engineers often prioritize using the Alternating Least Squares (ALS) technique. This technique optimizes alternately (fixes $X$ to solve the equation for $W$, then fixes $W$ to solve the equation for $X$). The ALS method transforms the problem into a convex form at each step, allowing for easy parallelization and accelerating convergence speeds on large data server clusters (Big Data).

**5. Evaluation: Pros and Cons**

Understanding the strengths and limitations of this model helps determine the appropriate use cases.
- **Pros:**
  - **Ultra-fast inference speed:** Once the model is trained, predicting a score is simply performing a matrix multiplication between $W$ and $X$, perfectly meeting the millisecond latency requirements of online systems.
  - **High performance with sparse data:** Thanks to the power of Low-Rank Approximation, the algorithm can delicately discover the connections between users and items even when the data is almost entirely empty.
  - **No Feature Engineering required:** The machine automatically learns the features (Latent Features) from user behavior, saving a massive amount of time extracting complex metadata as in Content-Based filtering.
- **Cons:**
  - **Lack of interpretability:** The latent feature vectors are just mathematical numbers. It is completely impossible to explain in human language why the system made a specific recommendation.
  - **Cold-start problem:** Powerless against a brand-new user or a newly launched item because there is no interaction history for the algorithm to calculate errors and update weights. (Mitigated by Mean Normalization).
  - **Inability to leverage Side Information:** The original MF model struggles to integrate auxiliary information such as user demographics (age, gender) or item metadata (director, movie genre).

#### Code Example: Matrix Factorization Collaborative Filtering – MovieLens 100K

```python
import numpy as np
import pandas as pd
import time
from scipy.sparse import coo_matrix
```

```python
class MatrixFactorization:
    def __init__(self, n_users, n_items, n_factors=10, learning_rate=5.0, regularization=0.02, n_epochs=100):
        self.n_users = n_users
        self.n_items = n_items
        self.n_factors = n_factors
        self.learning_rate = learning_rate
        self.regularization = regularization
        self.n_epochs = n_epochs

        # Init Latent Factors
        scale = 1.0 / self.n_factors
        self.P = np.random.normal(scale=scale, size=(n_users, n_factors))
        self.Q = np.random.normal(scale=scale, size=(n_items, n_factors))

        # Init Biases
        self.bu = np.zeros(n_users)  # User bias
        self.bi = np.zeros(n_items)  # Item bias
        self.global_mean = 0

        # Save history
        self.train_rmse_history = []
        self.test_rmse_history = []

    def fit(self, train_data, test_data=None, verbose=True):
        print(f"Training Vectorized MF (Factors={self.n_factors}, LR={self.learning_rate}, Reg={self.regularization})...")
        self.global_mean = np.mean(train_data[:, 2])  # Average rating

        u_idx = train_data[:, 0].astype(int)
        i_idx = train_data[:, 1].astype(int)
        ratings = train_data[:, 2]

        weights = train_data[:, 5] if train_data.shape[1] > 5 else np.ones_like(ratings)
        n_samples = len(ratings)

        for epoch in range(self.n_epochs):
            start_time = time.time()

            # 1. Predict ratings
            dot_products = np.sum(self.P[u_idx] * self.Q[i_idx], axis=1)
            preds = self.global_mean + self.bu[u_idx] + self.bi[i_idx] + dot_products

            # 2. Compute errors
            errors = ratings - preds
            weighted_errors = weights * errors

            # 3. Sparse Matrix
            E = coo_matrix((weighted_errors, (u_idx, i_idx)), shape=(self.n_users, self.n_items))

            # 4. Compute gradients for biases
            grad_b_u = np.squeeze(np.asarray(E.sum(axis=1))) - self.regularization * self.bu
            grad_b_i = np.squeeze(np.asarray(E.sum(axis=0))) - self.regularization * self.bi

            # 5. Compute gradients for latent factors
            grad_P = E.dot(self.Q) - self.regularization * self.P
            grad_Q = E.T.dot(self.P) - self.regularization * self.Q

            # 6. Update parameters
            self.bu += self.learning_rate * (grad_b_u / n_samples)
            self.bi += self.learning_rate * (grad_b_i / n_samples)
            self.P += self.learning_rate * (grad_P / n_samples)
            self.Q += self.learning_rate * (grad_Q / n_samples)

            # Evaluate RMSE
            train_rmse = self.evaluate(train_data)
            self.train_rmse_history.append(train_rmse)

            log_msg = f"Epoch {epoch+1:03d}/{self.n_epochs} - Train RMSE: {train_rmse:.4f}"
            if test_data is not None:
                test_rmse = self.evaluate(test_data)
                self.test_rmse_history.append(test_rmse)
                log_msg += f" | Test RMSE: {test_rmse:.4f}"
            log_msg += f" | Time: {time.time() - start_time:.4f}s"
            if verbose:
                print(log_msg)

    def evaluate(self, data):
        u_idx = data[:, 0].astype(int)
        i_idx = data[:, 1].astype(int)
        ratings = data[:, 2]

        dot_products = np.sum(self.P[u_idx] * self.Q[i_idx], axis=1)
        preds = self.global_mean + self.bu[u_idx] + self.bi[i_idx] + dot_products

        preds = np.clip(preds, 1, 5)

        rmse = np.sqrt(np.mean((ratings - preds) ** 2))
        return rmse

    def predict(self, u, i):
        if u >= self.n_users or i >= self.n_items:
            return self.global_mean
        
        pred = self.global_mean + self.bu[u] + self.bi[i] + np.dot(self.P[u], self.Q[i])
        return pred
    
    def recommend(self, user_id, top_k=5):
        if user_id >= self.n_users:
            return []
        scores = self.global_mean + self.bu[user_id] + self.bi + np.dot(self.P[user_id], self.Q.T)
        top_prod_indices = np.argsort(scores)[-top_k:][::-1]
        return top_prod_indices
    
```

```python
r_cols = ["user_id", "movie_id", "rating", "unix_timestamp"]
ratings_base = pd.read_csv("data/ml-100k/ub.base", sep="\t", names=r_cols, encoding="latin-1")
ratings_test = pd.read_csv("data/ml-100k/ub.test", sep="\t", names=r_cols, encoding="latin-1")

rate_train = ratings_base.values
rate_test = ratings_test.values

rate_train[:, :2] -= 1
rate_test[:, :2] -= 1

n_users = int(max(np.max(rate_train[:, 0]), np.max(rate_test[:, 0]))) + 1
n_items = int(max(np.max(rate_train[:, 1]), np.max(rate_test[:, 1]))) + 1
print(f"Sum of Users: {n_users}, Sum of Movies: {n_items}")

mf = MatrixFactorization(n_users, n_items, n_factors=20, learning_rate=15.0, regularization=0.1, n_epochs=100)
mf.fit(rate_train, rate_test, verbose=True)

user_id = 10
recommendations = mf.recommend(user_id, top_k=5)
print(f"\nTop 5 recommendations for User {user_id + 1}: {recommendations}")
```

### Deep Learning for Content-Based Filtering (Two-Tower Architecture)

**1. Model Construction**

The Deep Learning for Content-Based Filtering system (Two-Tower Model) is built upon two completely independent neural networks designed to learn the latent representations of users and items.
- **Step 1: Building the User Tower**
  
  The input is a feature vector $x_u$ containing the user's personal information (e.g., age, gender, country, past behavior). This vector is passed through multiple dense layers (Multi-Layer Perceptron) using non-linear activation functions (typically ReLU). The output of this tower is an embedding vector $v_u$ representing the user's preferences. Typically, this vector is passed through an L2 Normalization layer to scale its length to 1, enhancing training stability.

- **Step 2: Building the Item Tower**

  Similarly, the input is a feature vector $x_m$ of the item (e.g., release year, genre, director). This neural network processes $x_m$ through its own dense layers to output an embedding vector $v_m$ representing the item. A strict prerequisite is that the output layer of the Item Tower must have the same dimensionality as the output layer of the User Tower (e.g., both mapping to a 32-dimensional vector space). The vector $v_m$ is also commonly L2 normalized.

- **Step 3: Prediction Layer**

  Unlike traditional Content-Based Filtering, the bias term $b^{(j)}$ is usually removed from this architecture without degrading performance. The predicted score $\hat{y}^{(i,j)}$ is calculated by combining the outputs of the two towers via a Dot Product (or Cosine Similarity if L2 normalized):
  $$
  \hat{y}^{(i, j)} = v_{u}^{(j)} \cdot v_{m}^{(i)}
  $$

  (Note: If predicting probabilities for a binary classification task, this result can be passed through a Sigmoid function).

**2. Loss Function**

The model learns to tune the parameters of both towers simultaneously through a joint loss function, which evaluates how accurately $v_u$ and $v_m$ predict the actual ground-truth labels.
- **Case 1: Explicit Feedback (e.g., Star Rating Prediction)**

  This is treated as a regression problem. The standard loss function is **Mean Squared Error (MSE)**, which calculates the sum of squared errors between the predicted and actual scores. To prevent overfitting, an **L2 Regularization (Weight Decay)** term is fully integrated, penalizing the weights across all hidden layers ($l$) of both the User Tower ($L_u$) and the Item Tower ($L_m$):
  $$
  J = \frac{1}{2} \sum_{(i, j):r(i, j)=1} \left(v_{u}^{(j)} \cdot v_{m}^{(i)} - y^{(i, j)} \right)^{2} + \frac{\lambda}{2} \sum_{l=1}^{L_{u}} \|W_{u}^{[l]}\|_{F}^{2} + \frac{\lambda}{2} \sum_{l=1}^{L_{m}} \|W_{m}^{[l]}\|_{F}^{2}
  $$

- **Case 2: Implicit Feedback at Massive Scale (Retrieval Stage)**

  Instead of using standard Binary Cross-Entropy, modern Two-Tower systems typically utilize **Sampled Softmax Loss (Contrastive Loss)** combined with In-batch Negative Sampling. The goal is to pull the user vector $v_u$ closer to the positive item vector $v_m^+$ while pushing it away from $K$ negative item vectors $v_{m, k}^-$. To prevent the neural network's weights from exploding to artificially minimize the loss, and to mitigate overfitting, an **L2 Regularization (Weight Decay)** term is strictly integrated. This term penalizes the matrices of weights ($W$) across all hidden layers ($l$) of both the User Tower ($L_u$) and the Item Tower ($L_m$). The complete loss function is defined as:
  $$
  J = - \sum \log \frac{\exp(v_u \cdot v_m^+)}{\exp(v_u \cdot v_m^+) + \sum_{k=1}^K \exp(v_u \cdot v_{m, k}^-)} + \frac{\lambda}{2} \sum_{l=1}^{L_u} ||W_u^{[l]}||_F^2 + \frac{\lambda}{2} \sum_{l=1}^{L_m} ||W_m^{[l]}||_F^2
  $$

**3. Optimization**

To optimize all parameters (weights and biases of the dense layers in both towers), the Mini-batch Gradient Descent algorithm is utilized in conjunction with advanced optimizers like Adam or Adagrad.

This process does not rely on manually derived update formulas. Instead, it leverages the Computation Graph of frameworks like TensorFlow (using tf.GradientTape()). The algorithm calculates the partial derivatives of the cost function $J$ with respect to every weight in both networks simultaneously, then updates them in the opposite direction of the gradient to find the global minimum.

**Crucial Optimization at Inference:** To solve latency issues in production environments, the item vectors $v_m$ for millions of products are **pre-computed** by the Item Tower and stored in a vector database. When a user visits the platform, the system only runs a forward pass on the User Tower once to obtain $v_u$. It then uses an Approximate Nearest Neighbor (ANN) search algorithm against the stored $v_m$ vectors to return the best matches within milliseconds.

**4. Evaluation: Pros and Cons**

Understanding the strengths and limitations of the Two-Tower architecture is essential for positioning it correctly within a recommender system pipeline.

**Pros:**
- **Completely Solves the Cold-Start Problem:** Both new items and new users can be accurately predicted and recommended immediately based on their features, without waiting for historical interaction (rating) data.
- **Superb Scalability:** Thanks to the architectural separation of the two towers and the ability to pre-compute item vectors, this model is the industry gold standard for the Retrieval stage across catalogs of tens of millions of items.
- **Non-linear Power:** Unlike basic Content-Based filtering, deep non-linear hidden layers empower the neural network to learn much more profound and complex representations from raw data.

**Cons:**
- **Lack of Feature Crossing:** Because the two towers remain completely isolated and only interact at the final dot product, the model struggles to learn deep, cross-feature interactions between users and items (this is exactly why a Ranking model is required downstream).
- **Black-Box Nature:** Unlike traditional Content-Based systems, the 32-dimensional embedding vectors $v_u$ and $v_m$ are abstract mathematical values. The system cannot explain in natural language why a specific recommendation was made.
- **Data Hungry:** Deep neural networks require a massive and diverse amount of training data for the embedding layers to converge accurately without overfitting.

#### Code example: Deep Learning for Content-Based Filtering (Two-Tower Architecture)

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dot, Input, UnitNormalization
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.regularizers import l2
```

##### DATA PREPARATION

```python
np.random.seed(42)
tf.random.set_seed(42)

num_users = 1000
num_items = 500
num_interactions = 20000

# User (7 features): Age (1) + Gender (1) + Occupation (5 one-hot)
user_features_matrix = np.random.rand(num_users, 7)
# Item (11 features): Release Year (1) + Genre (10 multi-hot)
item_features_matrix = np.random.rand(num_items, 11)

# Randomly generate interacted User-Item pairs
user_indices = np.random.randint(0, num_users, num_interactions)
item_indices = np.random.randint(0, num_items, num_interactions)

# Map indices to their corresponding feature vectors to create the Training set
X_train_user = user_features_matrix[user_indices]
X_train_item = item_features_matrix[item_indices]

# Simulate ratings from 1 to 5 stars
ratings = np.random.randint(1, 6, num_interactions)

y_train_scaled = (ratings - 1.0) / 4.0
```

##### BUILDING THE TWO-TOWER MODEL

```python
reg_lambda = 0.01

user_tower = Sequential([
    Dense(128, activation="relu", kernel_regularizer=l2(reg_lambda)),
    Dense(64, activation="relu", kernel_regularizer=l2(reg_lambda)),
    Dense(32, kernel_regularizer=l2(reg_lambda), name="user_embedding"),
], name="User_Tower_Model")

item_tower = Sequential([
    Dense(128, activation="relu", kernel_regularizer=l2(reg_lambda)),
    Dense(64, activation="relu", kernel_regularizer=l2(reg_lambda)),
    Dense(32, kernel_regularizer=l2(reg_lambda), name="item_embedding")
], name="Item_Tower_Model")

input_user = Input(shape=(7,), name="user_input")
input_item = Input(shape=(11,), name="item_input")

vu = user_tower(input_user)
vi = item_tower(input_item)

vu = UnitNormalization(axis=1)(vu)
vi = UnitNormalization(axis=1)(vi)

output = Dot(axes=1, name="dot_product")([vu, vi])

model = Model(inputs=[input_user, input_item], outputs=output)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.005), loss="mse", metrics=["mae"])
```

```python
model.summary(expand_nested=True)
```

##### Training the model

```python
history = model.fit(
    x=[X_train_user, X_train_item],
    y=y_train_scaled,
    batch_size=256,
    epochs=20,
    validation_split=0.2,
    verbose=1
)
```

##### Inference and inverse scaling

```python
# Take the first 5 pairs for a test prediction
sample_user_features = X_train_user[:5]
sample_item_features = X_train_item[:5]

predictions_scaled = model.predict([sample_user_features, sample_item_features])

predictions = predictions_scaled * 4.0 + 1.0
actual_ratings = ratings[:5]

print("\n--- SAMPLE PREDICTION RESULTS ---")
for i in range(5):
    print(f"Actual: {actual_ratings[i]} stars | Predicted: {predictions[i][0]:.2f} stars")
```
