## What are Recommender Systems
- Recommender Systems are tools that suggest items to users based on their behaviour, preferences or past interactions. They help users find relevant products, movies, songs or content without manually searching for them. They are widely used in platforms like Youtube, Amazon and Netflix.
    - Suggest items automatically
    - Use user behaviour and patterns.
    - Improve content discovery
    - Common in e-commerce and entertainment apps

## Approaches in Recommender Systems

For the systems to understand and provide recommendations, there are several different approaches, evolving from basic algorithm to complex neural network architectures. Below are the most common categories of models:

**1. Content-Based Filtering**

This model focuses on analyzing the features of the item itself and user's historical profile.
- **How it works:** It extracts item information (e.g., genre, director, descriptive keywords) into vectors. When a user positively interacts with an item, the system finds other items with similar feature vectors to recommend.
- **Pros:** Handles the "new item" problem well, as it only needs metadata to analyze. It operates independently without needing to collect data from other users.
- **Cons:** Highly prone to the "filter bubble" effect. The system tends to suggest content that is too safe and familiar, lacking the ability to broaden the user's interests.

**2. Collaborative Filtering (CF)**

In contrast to Content-Based Filtering, this method completely ignores the item's internal content. Instead, it forcuses on exploiting the wisdom of the crowd through user interaction behaviour (sush as clicks, ratings, and purchases).
- **User-based CF:** Finds users with similar interaction histories. If User A and User B share similar tastes, the system will recommend an item that A just liked to B.
- **Item-based CF:** Provides recommendations based on items that are frequently consumed together by many different users.
- **Pros:** Brings serendipity, helping users discover new and interesting items they might not have actively thought of.
- **Cons:** Faces a significant hurdle with the "cold-start" problem. The system struggles completely when dealing with a brand-new user or a product that no one has interacted with yet.

**3. Matrix Factorization**

This is a core techniques used to solve the problem of massive but highly sparse User-Item interaction matrices found in the real world.
- **How it works:** It uses linear algebra to decompose the original, large matrix into two smaller, lower-dimensional matrices representing the User and the Item. This process helps discover "latent features" that humans cannot directly define or observe.
- **Highlights:** Offers fast computation and the ability to accurately predict the empty cells (missing ratings) in the matrix. This is the foundational algorithm that drove the success of the famous Netflix Prize competition.

**4. Deep Learning Models**

When the data scale reaches millions of users and billions of interactions, Deep Learning models are applied to learn highly complex, non-linear relationships.
- **Neural Collaborative Filtering (NCF):** Replaces basic matrix multiplication (dot product) with neural network layers (like Multi-Layer Perceptrons), allowing the model to be much more flexible in learning representations.
- **Two-Tower Model:**  An architecture consisting of two independent neural network branches used to embed User and Item information into the same vector space. This is the industry standard for the "retrieval" stage at a massive scale.
- **Sequential Models:** Treats the user's interaction history as a sequence of events over time. It applies architectures like RNNs or Transformers to predict the user's next action based on their current context.

## Recommender Systems Models

### Content-Based Filtering
**1. Model Construction**

A Content-Based Filtering system is built on two main pillars: the Item Profile and the User Profile.

- **Step 1:** Feature Extraction to create the Item Profile.

  Each item must be transformed into a mathematical vector $x^{(i)}$. In practice, items do not come with pre-calculated scores like a romance level of 0.9. These features must be constructed from raw data such as metadata, text, or images. Particularly for systems recommending articles or documents, applying text mining techniques is essential. Common methods include using **TF-IDF** or **CountVectorizer** to turn product descriptions into vocabulary vectors, utilizing **Word Embeddings** like **Word2Vec** or **BERT** to capture deeper semantic meaning, and applying one-hot encoding for categorical features like genre, director, or producer.

- **Step 2:** Building the User Profile.

  Each user is represented by a parameter vector, such as $w^{(j)}$, and a bias $b^{(j)}$. This vector has the same dimensionality as the item profile. It represents the weight or level of importance that user $j$ places on each specific feature of the items.

- **Step 3:** Prediction.

  The predicted score $\hat{y}^{(i,j)}$ that user $j$ would give to item $i$ is calculated as the dot product of the user's preference vector and the item's feature vector:$$\hat{y}^{(i,j)} = w^{(j)} \cdot x^{(i)} + b^{(j)}$$

**2. Loss Function**

The choice of loss function depends on the type of user feedback data available.

- **Case 1:** Explicit Feedback (e.g., 1-5 star ratings)

  This is treated as a regression problem. A suitable loss function is Mean Squared Error (MSE) combined with L2 Regularization to prevent overfitting and keep the weights $w$ from becoming too large. The loss function for user $j$ is:
  $$
  J(w^{(j)}, b^{(j)}) = \frac{1}{2} \sum_{i:r(i, j)=1} \left(w^{(j)} \cdot x^{(i)} + b^{(j)} - y^{(i, j)} \right)^2 + \frac{\lambda}{2} \sum_{k=1}^{n}\left(w_{k}^{(j)} \right)^2
  $$

- **Case 2:** Implicit Feedback (e.g., click/no click, buy/no buy)
  
  In this scenario, the task becomes a binary classification problem. The predicted score is passed through a Sigmoid function $\sigma(z) = \frac{1}{1+e^{-z}}$, and the loss function used is Binary Cross-Entropy (Log Loss):
  $$
  J(w^{(j)}, b^{(j)}) = - \sum_{i:r(i, j)=1} \left[y^{(i, j)}\log(\hat{y}^{(i, j)}) + (1 - \hat{y}^{(i, j)})\log(1 - \hat{y}^{(i, j)}) \right] + \frac{1}{2}\sum_{k=1}^{n}\left(w_{k}^{(j)} \right)^2
  $$

**3. Optimization**

  To find the weights $w^{(j)}$ and bias $b^{(j)}$ that minimize the loss function, Gradient Descent is commonly used. The algorithm calculates the partial derivative of the loss function with respect to each parameter and updates them iteratively. The weight update rule for the regression problem is:
  $$
  w_{k}^{(j)} := w_{k}^{(j)} - \alpha \left(\sum_{i:r(i, j)=1} \left(\hat{y}^{(i, j)} - y^{(i, j)}\right) x_{k}^{(i)} + \lambda w_{k}^{(j)}\right)
  $$
  Here, $\alpha$ represents the learning rate. In production environments, engineers often rely on more advanced optimizers like **Adam** or **RMSprop** for deep learning models, or direct solvers like L-BFGS for faster convergence, rather than implementing basic Gradient Descent from scratch.

**4. Evaluation:** Pros and Cons

Understanding the strengths and limitations of this model helps determine the appropriate use cases.

- Pros:
  - No item cold-start problem: When a new item is added to the system and its features are extracted, it can be recommended immediately to relevant users without needing prior interactions.
  - User independence: The algorithm learns solely from a specific user's history and does not require data from other users on the platform.
  - High interpretability: It is straightforward to explain recommendations, such as informing a user that an article was suggested because they frequently read about certain topics.
- Cons:
  - Heavy reliance on feature engineering: The model's accuracy depends entirely on the quality of feature extraction. If only basic metadata is available, the recommendations will not be highly relevant.
  - Filter bubble effect: The system tends to suggest items that are very similar to what the user has already consumed. It lacks the ability to introduce entirely new genres or categories, limiting serendipity.
  - User cold-start problem: If a new user registers without any interaction history, the model cannot build a profile and is unable to generate recommendations.

#### Code Example: Content-Based Filtering – MovieLens 100K

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import Ridge
from math import sqrt
```

```python
# Load data
def load_data(path="data/ml-100k"):

    # Load user data
    u_cols = ["user_id", "age", "sex", "occupation", "zip_code"]
    users = pd.read_csv(f"{path}/u.user", sep="|", names=u_cols, encoding="latin-1")
    n_users = users.shape[0]

    # Load ratings data
    r_cols = ["user_id", "movie_id", "rating", "unix_timestamp"]
    ratings_base = pd.read_csv(f"{path}/ua.base", sep="\t", names=r_cols, encoding="latin-1")
    ratings_test = pd.read_csv(f"{path}/ua.test", sep="\t", names=r_cols, encoding="latin-1")

    rate_train = ratings_base.values
    rate_test = ratings_test.values

    # Load item data
    i_cols = ['movie_id', 'movie_title', 'release_date', 'video_release_date',
              'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation',
              "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama',
              'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
              'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
    items = pd.read_csv(f"{path}/u.item", sep="|", names=i_cols, encoding="latin-1")
    n_items = items.shape[0]

    return users, items, rate_train, rate_test, n_users, n_items
```

```python
# Build item profiles (TF-IDF)
def build_item_profiles(items):
    X0 = items.values
    gerne_matrix = X0[:, -19:]

    transformer = TfidfTransformer(smooth_idf=True, norm="l2")
    item_profiles = transformer.fit_transform(gerne_matrix.tolist()).toarray()
    return item_profiles
```

```python
# Get items rated by user
def get_items_rated_by_user(rate_matrix, user_id):
    y = rate_matrix[:, 0]
    ids = np.where(y == user_id + 1)[0]
    item_ids = rate_matrix[ids, 1] - 1
    scores = rate_matrix[ids, 2]
    return item_ids.astype(int), scores
    
```

```python
# Train content-based model
def train_content_based_model(item_profiles, rate_train, n_users):
    d = item_profiles.shape[1]
    W = np.zeros((d, n_users))
    b = np.zeros((1, n_users))

    for n in range(n_users):
        ids, scores = get_items_rated_by_user(rate_train, n)
        clf = Ridge(alpha=0.1, fit_intercept=True)

        Xhat = item_profiles[ids, :]
        clf.fit(Xhat, scores)

        W[:, n] = clf.coef_
        b[0, n] = clf.intercept_

        if n % 100 == 0:
            print(f"Trained model for user {n}/{n_users}")
    
    return W, b
```

```python
# Predict ratings for user
def predict(item_profiles, W, b):
    return item_profiles @ W + b
```

```python
# Evaluate model
def evaluate(Yhat, rates, n_users):
    se = 0
    cnt = 0

    for n in range(n_users):
        ids, scores_truth = get_items_rated_by_user(rates, n)
        scores_pred = Yhat[ids, n]

        e = scores_truth  - scores_pred
        se += np.sum(e ** 2)
        cnt += len(ids)
    return np.sqrt(se / cnt)
```

```python
# Recommend function
def recommend_top_k(user_id, Yhat, rate_train, items, k=5   ):
    rated_items, _ = get_items_rated_by_user(rate_train, user_id)

    scores = Yhat[:, user_id]
    scores[rated_items] = -np.inf

    top_ids = np.argsort(scores)[-k:][::-1]

    print(f"\nTop {k} recommendations for User {user_id + 1}:")

    for idx in top_ids:
        print(f"- {items.iloc[idx]['movie_title']} (Predicted: {scores[idx]:.2f})")
```

```python
users, items, rate_train, rate_test, n_users, n_items = load_data()

print("Number of users:", n_users)
print("Number of items:", n_items)

tfidf = build_item_profiles(items)

W, b = train_content_based_model(tfidf, rate_train, n_users)

Yhat = predict(tfidf, W, b)

print("\nRMSE Train:", evaluate(Yhat, rate_train, n_users))
print("RMSE Test :", evaluate(Yhat, rate_test, n_users))

recommend_top_k(user_id=10, Yhat=Yhat, rate_train=rate_train, items=items, k=5)
```
